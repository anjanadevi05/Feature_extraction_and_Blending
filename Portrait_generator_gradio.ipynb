```python
import gradio as gr
import cv2
import numpy as np

# Load the face cascade classifier
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
def extract_features(image, face_cascade):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5)
    return faces

def replace_features(img_target, img_source, features_target, features_source):
    result = img_target.copy()

    for (x_t, y_t, w_t, h_t), (x_s, y_s, w_s, h_s) in zip(features_target, features_source):
        # Extract the regions from the source and target images
        region_target = img_target[y_t:y_t + h_t, x_t:x_t + w_t]
        region_source = img_source[y_s:y_s + h_s, x_s:x_s + w_s]

        # Resize the source region to match the size of the target region
        region_source = cv2.resize(region_source, (w_t, h_t))

        # Create a mask based on the contours of the face
        mask = np.zeros_like(region_target)
        cv2.ellipse(mask, (w_t // 2, h_t // 2), (w_t // 2, h_t // 2), 0, 0, 360, (255, 255, 255), -1)

        # Blend or replace regions with mask-based blending
        blended_region = cv2.seamlessClone(region_source, region_target, mask, (w_t // 2, h_t // 2), cv2.NORMAL_CLONE)

        # Replace the region in the result
        result[y_t:y_t + h_t, x_t:x_t + w_t] = blended_region

    return result



def face_replacement(input_image):
    # Load the source image
    img_source = cv2.imread("/content/orginal.jpg")

    # Extract facial features from the target image
    features_target = extract_features(input_image, face_cascade)
    features_source = extract_features(img_source, face_cascade)

    # Replace facial features with those from the source image
    result_image = replace_features(input_image, img_source, features_target, features_source)
    result_gray = cv2.cvtColor(result_image, cv2.COLOR_BGR2GRAY)

    return result_gray

# Create a Gradio interface
iface = gr.Interface(
    fn=face_replacement,
    inputs="image",
    outputs="image",
    live=True,
)

# Launch the Gradio interface
iface.launch()
```
